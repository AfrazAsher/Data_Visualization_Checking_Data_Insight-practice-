# I'm going to analyze the data using Data Visualization. We are going to use "Covit-19: Pakistan" as our Dataset.

Step-1: Importing all the important libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt #As a failsafe
```

Step-2: Importing the data-set into the file

```python
data = pd.read_csv(r"owid-covid-data-pakistan.csv")
```

Step-3: check whether Data-Set is imported successfully or not ?

```python
# Checking if there are Null values in the data
data.isnull().sum()
```

```python
sns.relplot(x="total_cases",y="total_deaths", data=data)
```

```python
sns.relplot(x="total_cases",y="total_tests", data=data)
```

```python
sns.relplot(x="total_cases",y="total_vaccinations", data=data)
```

```python
sns.relplot(x="total_vaccinations",y="total_boosters", data=data)
```

```python
sns.relplot(x="total_cases",y="total_deaths" ,hue="people_fully_vaccinated", data=data)
```

```python
sns.relplot(x="population",y="total_cases", data=data)
```

```python
sns.relplot(x="total_deaths",y="median_age", data=data)
```

```python
sns.pairplot(data, vars=["total_cases", "new_cases", "total_deaths", "new_deaths", "total_tests", "new_tests",
                         "total_vaccinations", "people_fully_vaccinated", "total_boosters"])
```

```python
sns.relplot(x="total_cases",y="tests_per_case", kind="line" ,data=data)
# we can specify the type of plot in seaborn useing the "kind" parameter
```

```python
sns.catplot(x="date", y="new_cases", data=data)
```

```python
plt.plot(data["new_cases"], label="Cases")
plt.plot(data["new_deaths"], label="Deaths")
plt.legend()
plt.title("Evolution of Covid-19 Cases")
plt.xlabel('Days')
plt.ylabel('Number of Patients')
plt.show()
```
